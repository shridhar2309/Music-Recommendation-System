```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

# Set up credentials
client_id = '3d627552a7404b4b9b4db42bd08581d5'
client_secret = '3a68d462cba64a898397ebb332811342'

# Authenticate with Spotify
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))
# Define a playlist ID (this is a sample; use any playlist ID you want)
playlist_id = '37i9dQZF1DXcBWIGoYBM5M'  # Spotify's "Today's Top Hits" playlist

results = sp.playlist_tracks(playlist_id)
track_ids = [track['track']['id'] for track in results['items']]

tracks_data = []
for track in results['items']:
    track_info = track['track']
    track_id = track_info['id']
    features = sp.audio_features(track_id)[0]  # Get features for each track
    if features:
        track_data = {
            'track_id': track_id,
            'danceability': features['danceability'],
            'energy': features['energy'],
            'tempo': features['tempo'],
            'loudness': features['loudness'],
            'valence': features['valence'],
            'popularity': track_info['popularity']  # Include popularity
        }
        tracks_data.append(track_data)

# Convert to DataFrame
df = pd.DataFrame(tracks_data)
df.to_csv('spotify_tracks.csv', index=False)
df = pd.read_csv('spotify_tracks.csv')

# List of required columns
required_columns = ['danceability', 'energy', 'tempo', 'loudness', 'valence']

# Add missing columns with default values
for col in required_columns:
    if col not in df.columns:
        df[col] = 0.5  # or np.nan, depending on your preference

# Drop rows with NaNs, if any
df = df.dropna(subset=required_columns)

# Ensure numeric data for relevant columns
for col in required_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Scale selected features
scaler = MinMaxScaler()
df[['loudness', 'tempo']] = scaler.fit_transform(df[['loudness', 'tempo']])

# Apply PCA for dimensionality reduction
pca = PCA(n_components=3)
df_pca = pca.fit_transform(df[required_columns])
print("PCA applied successfully.")
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the figure size
plt.figure(figsize=(15, 10))

# Plot each audio feature distribution
audio_features = ['danceability', 'energy', 'tempo', 'loudness', 'valence']
for i, feature in enumerate(audio_features, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[audio_features].corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of Audio Features")
plt.show()
sns.pairplot(df[audio_features])
plt.suptitle("Pair Plot of Audio Features", y=1.02)
plt.show()
from mpl_toolkits.mplot3d import Axes3D

# Assuming 'df_pca' is the PCA-transformed data with 3 components
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2', 'PC3'])

# 3D Plot for PCA components
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_pca['PC1'], df_pca['PC2'], df_pca['PC3'], c='skyblue', s=50)
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.title("3D PCA of Spotify Tracks")
plt.show()
# Define popularity categories
df['popularity_level'] = pd.cut(df['popularity'], bins=[0, 30, 60, 100], labels=['Low', 'Medium', 'High'])

# Plot feature distributions by popularity level
plt.figure(figsize=(15, 10))
for i, feature in enumerate(audio_features, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(data=df, x='popularity_level', y=feature)
    plt.title(f'{feature} by Popularity Level')

plt.tight_layout()
plt.show()
from sklearn.manifold import TSNE

# Use TSNE for dimensionality reduction to 2 components
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(df[audio_features])

# Plot t-SNE
plt.figure(figsize=(8, 6))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c='skyblue', s=10, alpha=0.7)
plt.title("t-SNE Plot of Spotify Tracks")
plt.xlabel("TSNE Component 1")
plt.ylabel("TSNE Component 2")
plt.show()
import pandas as pd
from sklearn.model_selection import train_test_split

# Load data
df = pd.read_csv('spotify_tracks.csv')

# Prepare features and target
features = df[['danceability', 'energy', 'tempo', 'loudness', 'valence']]
target = df['popularity']  # Use popularity as a target variable

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
from sklearn.neighbors import NearestNeighbors

# Create KNN model
knn = NearestNeighbors(n_neighbors=5)  # Set the number of neighbors
knn.fit(X_train)  # Fit the model to the training data
def recommend_tracks(track_index, n_neighbors=5):
    # Get the track features for which to find recommendations
    sample_features = X_test.iloc[track_index].values.reshape(1, -1)

    # Get indices of nearest neighbors
    distances, indices = knn.kneighbors(sample_features, n_neighbors=n_neighbors)

    # Retrieve recommended tracks using the indices
    recommended_tracks = df.iloc[indices[0]]

    return recommended_tracks[['track_id', 'popularity', 'danceability', 'energy', 'tempo', 'loudness', 'valence']]

# Example usage: Recommend tracks based on the first test sample
recommended_tracks = recommend_tracks(track_index=0)  # Change index as needed
print("Recommended Tracks:")
print(recommended_tracks)
# Example for evaluating the first test sample
def evaluate_recommendation(track_index):
    actual_popularity = y_test.iloc[track_index]
    recommended_tracks = recommend_tracks(track_index)

    # Calculate the average popularity of recommended tracks
    average_recommended_popularity = recommended_tracks['popularity'].mean()

    print(f'Actual Popularity: {actual_popularity}')
    print(f'Average Popularity of Recommended Tracks: {average_recommended_popularity}')

# Evaluate for the first test sample
evaluate_recommendation(track_index=0)  # Change index as needed

import streamlit as st

st.title("Music Recommendation System")
track_id = st.text_input("Enter a Track ID")

if st.button("Recommend"):
    recommended_tracks = recommend_tracks(track_id)
    st.write(recommended_tracks)
```
